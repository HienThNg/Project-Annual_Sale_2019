## Task 2: Clean and Preprocess data

### Task 2.1: Merge 12-month data

```python
os.listdir(path)
```

```python
frames = []
all_length = []
for file in os.listdir(path):
    if file.endswith('.csv'):
        filepath = path + file
        df1 = pd.read_csv(filepath)
        frames.append(df1)
        result = pd.concat(frames)
        length_1month = len(df1.index)
        all_length.append(length_1month)

df = result
df.to_csv('annualSales2019.csv', index=False)
print(df.shape)
df.head()
```

```python
print(sum(all_length))
```

### Task 2.2: Add 'Month' column

```python
df['Month'] = df['Order Date'].str[0:2]
df.head()
```

```python
print(set(df['Month']))
```

### task 2.3: Get rid of 'NaN' and 'Or' value

```python
df = df.dropna(how='all')
df.head()
```

```python
df = df[df['Month'] != 'Or']
df.head()
```

```python
print(set(df['Month']))
```

### Task 3.3: What time should we display ads to maximize the likelihood of customer's buying product?

```python
df.head()
```

```python
print(df['Order Date'].dtypes)
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
print(df['Order Date'].dtypes)
```

```python
df['Hours'] = df['Order Date'].dt.hour
```

```python
df.groupby('City').sum()
```

**The total of money earned** 

```python
sales_value_hours = df.groupby('Hours').sum()['Sales']
sales_value_hours.max()
```

```python
hours = []
for hour, sales in sales_value_hours.items():
    hours.append(hour)
print(hours)
```

```python
plt.plot(hours, sales_value_hours)
plt.grid()
plt.xticks(hours, size=8)
plt.xlabel('Hours')
plt.ylabel('Sales in USD')
plt.show()
```

**The total of order**

```python
sales_value_hours = df.groupby('Hours').count()['Sales']

hours = []
for hour, sales in sales_value_hours.items():
    hours.append(hour)

plt.plot(hours, sales_value_hours)
plt.grid()
plt.xticks(hours, size=8)
plt.xlabel('Hours')
plt.ylabel('Sales in USD')
plt.show()
```

**Comments**

- People was likely to buy at 11-12am (noon break time) and 6-8pm (after work).
- The company should run promotions from 30 to 60 minutes before the peak hours.
